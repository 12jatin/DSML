```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing


california_housing = fetch_california_housing(as_frame=True)
data = california_housing.frame


print(data.head())
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.5)
plt.title('Scatter Plot of Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.5)
plt.title('Scatter Plot of Longitude vs Latitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
data['MedInc'].hist(bins=30)
plt.title('Histogram of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data.boxplot(column='MedInc')
plt.title('Box Plot of Median Income')
plt.ylabel('Median Income')
plt.show()
```

```python
data.groupby('MedInc')['AveRooms'].mean().plot(kind='line', marker='o')
plt.title('Line Plot of Median Income by Median income')
plt.xlabel('AveRooms')
plt.ylabel('Median Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.pointplot(x=data.index, y=data['MedInc'])
plt.title('Point Plot')
plt.xlabel('Sample Index')
plt.ylabel('Median Income')
plt.show()
```

```python
sns.jointplot(x='MedInc', y='HouseAge', data=data, kind='scatter')
plt.title('Joint Plot of Median Income vs House Age')
plt.show()
```

```python
sns.regplot(x='MedInc', y='HouseAge', data=data)
plt.title('Regplot')
plt.show()
```
